```python
#Dependencies
import numpy as np
import pandas as pd
```

```python
#Load both CSV files
loaded_schools = "Resources/schools_complete.csv"
loaded_students = "Resources/students_complete.csv"
```

```python
#Read school and student files and store into Pandas DataFrames
schools_data = pd.read_csv(loaded_schools)
students_data = pd.read_csv(loaded_students)
```

```python
#Combine de data files into a single dataset
complete_schools_data = pd.merge(students_data, schools_data, how="left", on=["school_name", "school_name"])
complete_schools_data.head()
```

```python
#Verify no info. is missing
complete_schools_data.count()
```

```python
#Rename column headers so they are easier to read
complete_schools_data = complete_schools_data.rename(columns={"student_name":"Student Name", "gender":"Gender",
                                                              "grade": "Grade", "school_name": "School Name",
                                                              "reading_score": "Reading Score", "math_score": "Math Score",
                                                             "type":"Type", "size": "Size", "budget": "Budget"})
complete_schools_data.head()
```

```python
#Reorganize columns in an order that makes more sense to ease review of info.
complete_schools_data = complete_schools_data[["School Name", "School ID", "Type", "Size", "Budget", "Student Name", "Student ID",
                                               "Gender", "Grade", "Reading Score", "Math Score"]]
complete_schools_data.head()
```

```python
#BEGIN DISTRICT SUMMARY OF HW
#Count the unique number of schools in the dataset to calculate the total # of schools in the district
count_unique_schools = complete_schools_data["School Name"].nunique()
count_unique_schools
```

```python
#Separate the "School Name" and "Budget" columns to help analyze the district's total budget
district_budget_data = complete_schools_data.loc[:, ["School Name", "Budget"]]
district_budget_data
```

```python
#Eliminate the duplicates from these 2 columns to get an accurate district budget total in the following step
total_unique_district_data = district_budget_data.drop_duplicates(subset="School Name", keep="first")
total_unique_district_data
```

```python
#Calculate the total budget for the district
total_budget_data = total_unique_district_data["Budget"].sum()
total_budget_data
```

```python
#Count the unique number of students in the district
total_unique_students = complete_schools_data["Student ID"].nunique()
total_unique_students
```

```python
#Separate "Student Name" and "Math Score" columns to begin average math score prompt
student_math_data = complete_schools_data.loc[:, ["Student Name", "Math Score"]]
student_math_data
```

```python
#Calculate the average math score in the district
avg_math_score = student_math_data["Math Score"].mean()
avg_math_score
```

```python
#Separate "Student Name" and "Reading Score" columns to begin average reading score prompt
student_reading_data = complete_schools_data.loc[:, ["Student Name", "Reading Score"]]
student_reading_data
```

```python
#Calculate the average reading score in the district
avg_reading_score = student_reading_data["Reading Score"].mean()
avg_reading_score
```

```python
#Filter by those students with >= 70 math scores to begin % passing math prompt
passing_math_data = complete_schools_data.loc[complete_schools_data["Math Score"] >= 70, :]
passing_math_data
```

```python
#Count the number of students with math scores >= 70
passing_math = passing_math_data.shape[0]
passing_math
```

```python
#Filter by those students with >= 70 reading scores to begin % passing reading prompt
passing_reading_data = complete_schools_data.loc[complete_schools_data["Reading Score"] >= 70, :]
passing_reading_data
```

```python
#Count the number of students with reading scores >= 70
passing_reading = passing_reading_data.shape[0]
passing_reading
```

```python
#Calculate the percent of students with >= 70 math scores
percent_passing_math = (passing_math/total_unique_students)*100
percent_passing_math
```

```python
#Calculate the percent of students with >= 70 reading scores
percent_passing_reading = (passing_reading/total_unique_students)*100
percent_passing_reading
```

```python
#Separate out all those students with >= 70 in both math and reading scores
overall_passing_data = complete_schools_data.loc[(complete_schools_data["Math Score"] >= 70) &
                                                 (complete_schools_data["Reading Score"] >= 70), :]
overall_passing_data
```

```python
#Count the number of students in this new dataset
overall_passing_students = overall_passing_data.shape[0]
overall_passing_students
```

```python
#Calculate the overall percent of students who have passing scores in both math and reading
percent_overall_passing = (overall_passing_students/total_unique_students)*100
percent_overall_passing
```

```python
#BEGIN SCHOOL SPECIFIC SUMMARY OF HW
#Create a whole new DataFrame using the original DataFrame for this section of the HW
complete_schools_data2 = pd.DataFrame(complete_schools_data)
complete_schools_data2
```

```python
#Organize the DataFrame by schools
complete_schools_data2 = complete_schools_data2.sort_values("School Name")
complete_schools_data2.head()
```

```python
#Remove the "School ID" column from DataFrame
complete_schools_data2 = complete_schools_data2.drop("School ID", axis=1)
complete_schools_data2
```

```python
#Group the DataFrame by "School Name"
groupby_complete_schools2 = complete_schools_data2.groupby(["School Name"])
groupby_complete_schools2.head(15)
```

```python
#Group the DataFrame by "Student Name" to get the total students per school count
total_students_per_school = groupby_complete_schools2["Student Name"].count()
total_students_per_school.head()
```

```python
groupby_math = complete_schools_data2.groupby(["School Name"])["Math Score"].mean()
groupby_math.head()
```

```python
groupby_reading = complete_schools_data2.groupby(["School Name"])["Reading Score"].mean()
groupby_reading.head()
```

```python
#Calculate the number of students who passed Math per school
math_passing_data_per_school = complete_schools_data2.loc[(complete_schools_data2["Math Score"] >= 70), :]

math_passing_data_per_school = math_passing_data_per_school.groupby("School Name")

math_pass = math_passing_data_per_school["Math Score"].count()

math_pass
```

```python
#Calculate the "% Passing Math" per school
percent_math_pass = (math_pass/total_students_per_school)*100
percent_math_pass
```

```python
#Calculate the number of students who passed Reading per school
reading_passing_data_per_school = complete_schools_data2.loc[(complete_schools_data2["Reading Score"] >= 70), :]

reading_passing_data_per_school = reading_passing_data_per_school.groupby("School Name")

reading_pass = reading_passing_data_per_school["Reading Score"].count()

reading_pass
```

```python
#Calculate the "% Passing Reading" per school
percent_reading_pass = (reading_pass/total_students_per_school)*100
percent_reading_pass
```

```python
#Calculate the number of students who passed Reading and Math per school 
overall_passing_data2 = complete_schools_data2.loc[(complete_schools_data2["Math Score"] >= 70) &
                                                 (complete_schools_data2["Reading Score"] >= 70), :]

overall_passing_data2 = overall_passing_data2.groupby("School Name")

overall_pass = overall_passing_data2["Student Name"].count()

overall_pass
```

```python
#Calculate the "% Overall Passing" per school
percent_overall_pass = (overall_pass/total_students_per_school)*100
percent_overall_pass
```

```python
#Create a new DataFrame to hold the results for the "School Summary" section of the HW
school_summary_df = pd.DataFrame(complete_schools_data2)
school_summary_df
```

```python
#Remove duplicate schools
school_summary_df = school_summary_df.drop_duplicates(subset="School Name", keep="first")
school_summary_df
```

```python
#Make the "School Name" column into the index
school_summary_df = school_summary_df.set_index("School Name", drop=True)
school_summary_df
```

```python
#Rename School Type and Total School Budget columns
school_summary_df1 = school_summary_df.rename(columns = {"Type":"School Type", "Budget":"Total School Budget"})
school_summary_df1
```

```python
#Delete unnecessary columns from DataFrame
school_summary_df1 = school_summary_df1.drop(["Size", "Student Name", "Student ID", "Gender", "Grade",
                                            "Reading Score", "Math Score"], axis=1)
school_summary_df1
```

```python
#Add a "Total Students" column
school_summary_df1["Total Students"] = total_students_per_school
school_summary_df1
```

```python
#Reorganize Columns
school_summary_df1 = school_summary_df1[["School Type", "Total Students", "Total School Budget"]]
school_summary_df1
```

```python
#Calculate and add "Per Student Budget" column
for ind, row in school_summary_df1.iterrows():
    school_summary_df1.loc[ind, "Per Student Budget"] = (row["Total School Budget"])/(row["Total Students"])
school_summary_df1
```

```python
#Add "Average Math Score" column
school_summary_df1["Average Math Score"] = groupby_math
school_summary_df1
```

```python
#Add "Average Reading Score" column
school_summary_df1["Average Reading Score"] = groupby_reading
school_summary_df1
```

```python
#Add "% Passing Math" column
school_summary_df1["% Passing Math"] = percent_math_pass
school_summary_df1
```

```python
#Add "% Passing Reading" column
school_summary_df1["% Passing Reading"] = percent_reading_pass
school_summary_df1
```

```python
#Add "% Overall Passing" column and print final DataFrame for School Summary Section
school_summary_df1["% Overall Passing"] = percent_overall_pass
school_summary_df1
```

```python
#Sort and display the top five performing schools by % overall passing
top_5_schools = school_summary_df1.sort_values("% Overall Passing", ascending=False)
top_5_schools = top_5_schools.loc[top_5_schools["% Overall Passing"] > 90.50, :]
top_5_schools
```

```python
#Sort and display the bottom five performing schools by % overall passing
bottom_5_schools = school_summary_df1.sort_values("% Overall Passing")
bottom_5_schools = bottom_5_schools.loc[bottom_5_schools["% Overall Passing"] < 54, :]
bottom_5_schools
```
